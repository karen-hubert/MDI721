```python
from collections import Counter
c = Counter()
c.update(['la', 'fonction', 'findall', 'est', 'très', 'utile',
          'un', 'objet', 'de', 'type', 'counter', 'est', 'aussi', 'utile'])
c
```

```python
#le findall ne revient pas à un split
#split éclate la chaine de caractères 
'la fonction findall est très utile.'.split()
```

```python
#avec split, la ponctuation est prise en compte
#le split ne prend qu'un séparateur qui est défini par défaut sur l'espace. 
#re.findall est une capture positive et récupère tous les items composés de ces termes
```

```python
liste = [
    {'categorie': node_category.attrib['titre'],
    'sousCategorie': node_subcategory.attrib['titre'],
    'proposition': node_proposition.text}
        for node_category in root.getiterator('categorie')
            for node_subcategory in node_category.getiterator('sousCategorie')
                for node_proposition in node_subcategory.getiterator('proposition')]

df = pd.DataFrame(liste)
df
```

```python
#on a tranformé les propositions en liste de mots
df['proposition'].apply(lambda x : re.findall('[A-Za-zÀ-ÿ0-9]+',x.lower()))
```

```python
c = Counter()

df['proposition'].apply(lambda x : c.update(re.findall('[A-Za-zÀ-ÿ0-9]+',x.lower())))

#on s'intéresse à l'effet de bord i.e. au dictionnaire c et pas au résultat de la ligne précédente
c
```

```python
#comment mettre c dans une série ? 
#il faut donner les valeurs à transformer et un index
voc = pd.Series(list(c.values()), index = c.keys())
voc = voc.sort_values(ascending = False )
voc

# dictionnaire trié par ordre décroissant
#les stopwords arrivent en premier
#cf. librairie nltk.corpus pour enlever les stopwords
```
