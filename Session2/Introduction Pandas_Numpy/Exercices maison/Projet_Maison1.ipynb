```python
import numpy as np
import pandas as pd

geo = pd.read_csv("correspondance-code-insee-code-postal.csv",
                   sep=';')
geo
```

```python
#création de la liste comprenant le CP et la Ville. 

CP_ville = []
for CP, ville in zip(geo["Code Postal"], geo["Commune"]):
    CP_ville.append(str(CP) + " " + str(ville))
        
CP_ville
```

```python
# Intégrer une nouvelle colonne dans le dataframe
geo.insert(0,"CP Ville",CP_ville)
geo
```

```python
latitude = []
longitude = []

for point_2d in (geo["geo_point_2d"]):
    tab = point_2d.split(',')
    latitude.append(float(tab[0]))
    longitude.append(float(tab[1]))

geo_modif = geo
geo_modif.insert(10,"latitude",latitude)
geo_modif.insert(11,"longitude",longitude)

geo_modif
```

```python
#Ecrivez une fonction qui détermine la commune la plus proche d'un point à partir de sa latitude et sa longitude.
from math import *

def commune_plus_proche(latitude_input,longitude_input, geo_modif):
    #calcul de la distance angulaire 
    #Si l’on considère deux points A et B sur la sphère, 
    #de latitudes φA et φB et de longitudes λA et λB, 
    #alors la distance angulaire en radians SA-B entre A et B est donnée par 
    #la relation fondamentale de trigonométrie sphérique, utilisant dλ = λB – λA :
    #SA-B = arc cos (sin φA sin φB + cos φA cos φB cos dλ)
    #on calcule la distance angulaire pour toutes les communes et on retient la distance minimale
    
    if 'distance angulaire' in geo_modif: 
        del geo_modif['distance angulaire']
    
    else:
        d_angulaire = []
        for long, lat in zip(geo_modif["longitude"],geo_modif["latitude"]) :
            d_angulaire.append(acos(sin(lat) * sin(latitude_input) + cos(lat) * cos(latitude_input) * cos(long - longitude_input)))
    
        
        
    geo_modif.insert(10,"distance angulaire",d_angulaire)
    index_min = geo_modif['distance angulaire'].idxmin()   
    return geo_modif.iloc[index_min]

commune_plus_proche(44, 4.5, geo_modif) #ici on souhaite la commune la plus proche du point longitude 44 et latitude 4.5
```

```python
# Ajoutez une fonction de conversion pour pouvoir utiliser la première fonction avec un GPS
#(degrés, minutes, secondes).

def conversion_GPS_longlat(DMS_lat,DMS_long):
    #DMS_lat = 48°51'31.0''N / DMS_long = 2°17'40.2''E
    #les arguments de cette fonction sont des chaines de caractères 
    
    #interpretation des coordonnes
    
    ##partir entière 48 / 2
    ##minutes 51 / 17
    ##secondes 31.0 / 40.2
    ##orientation N / E
    
    #récupération de la partie entière
    index_degree_lat = DMS_lat.find('°')
    entiere_lat = int(DMS_lat[0: index_degree_lat])
    index_degree_long = DMS_long.find('°')
    entiere_long = int(DMS_long[0:index_degree_long])
    
    #récupération de l'orientation N/S ou O/E
    orientation_lat = DMS_lat[-1]
    orientation_long = DMS_long[-1]
    
    #Mise en place de l'orientation 
   
    if ((orientation_lat != 'N') or (orientation_long != 'E')): 
        entiere_lat = -1 * entiere_lat
        entiere_long = -1 * entiere_long
        
    
    #récupération des minutes
    index_min_lat = DMS_lat.find("'")
    min_lat = float(DMS_lat[index_degree_lat + 1: index_min_lat])/60
    index_min_long = DMS_long.find("'")
    min_long = float(DMS_long[index_degree_long+1:index_min_long])/60
    
    #récupération des secondes
    index_second_lat = DMS_lat.find("''")
    second_lat = float(DMS_lat[index_min_lat + 1: index_second_lat])/3000
    index_secon_long = DMS_long.find("''")
    second_long = float(DMS_long[index_min_long +1:index_secon_long])/3000
    
    longitude = entiere_long + min_long + second_long
    latitude = entiere_lat + min_lat + second_lat
   
    return {longitude, latitude}
    

conversion_GPS_longlat("48°51'31.0''N","2°17'40.2''E") 
#fonction qui convertit les données GPS en longitude latitude
```
