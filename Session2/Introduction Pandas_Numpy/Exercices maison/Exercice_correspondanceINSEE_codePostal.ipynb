```python
# import des modules usuels
import numpy as np
import pandas as pd

geo = pd.read_csv("correspondance-code-insee-code-postal.csv",
                   sep=';',
                   usecols=range(11))
```

```python
#ici création de  la liste l composées de "Commune", "Altitude Moyenne", "Superficie"
#et "Population"
l = ["Commune","Altitude Moyenne", "Superficie","Population"]
geo[l].head()
```

```python
#Question 1
#Affichez l'altitude moyenne, la superficie et la population 
#de toutes les communes dénommées Sainte-Colombe.

geo.loc[geo["Commune"] == "SAINTE-COLOMBE", "Altitude Moyenne" : "Population"]

#Attention, on ne peut pas faire 
#l.loc[l["Commune"] == "SAINTE-COLOMBE"]
# car l est une liste... et ne possède pas LOC
```

```python
#Combien y a-t-il de départements dans lesquels au moins une commune 
#a plus de 200.000 habitants ?

l = geo.loc[geo["Population"]>=0.2]  # ici on récupère toutes les lignes dont la population est > 0.2 (200_000)
liste_departement = [] #on crée une liste de départements vides et on rajoute les départements qui sont intégrés dans l (on a déjà fait le tri dans l)
for departement in l["Département"]:
    if departement not in liste_departement:
        liste_departement.append(departement)
        
print("Nombre de département dont la population dépasse les 200k:", len(liste_departement))
#geo.loc[geo["Population"]>=0.2]

#normalement le nombre de commune devrait être beaucoup plus important que le nombre de département
print("Nombre de Communes dont la population dépasse les 200k: ", len(l["Commune"]))
```

```python
#Combien y a-t-il de communes dont le nom comporte 2 lettres?
#Il faut mesurer pour chaque commune la longueur de son nom 
s = geo['Commune']
l = s.loc[s.str.len()==2]
print(l)
print ('Il y a',len(l), 'communes dont le nom comporte 2 lettres')
```

```python
#Donnez la liste de ces communes dans l'ordre alphabétique.

s = geo['Commune']
l = s.loc[s.str.len()==2]
#l.sort() ou sort(l)

#Possibilité plus longue
l2 = []
for commune in l: 
    l2.append(commune)
l2.sort()

print("la liste des communes dans l'ordre alphabétique :", l2)
```

```python
#Sélectionnez les communes pour lesquelles l'atitude moyenne vaut la superficie.


L3 = geo.loc[geo["Altitude Moyenne"]==geo["Superficie"]]  
L3

```

```python
#Quelle est la population totale de ces communes ?
print("La population totale de ces commmunes est:", L3["Population"].sum())
```
